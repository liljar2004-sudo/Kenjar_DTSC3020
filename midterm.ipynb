```python
def movie_ticket_booth():
    prices = {
        "adult": 12.5,
        "child": 8,
        "senior": 9.5,
        "student": 10
    }

    subtotal = 0.0

    while True:
        user_input = input("Enter ticket type (adult, child, senior, student) or 'done' to finish: ").strip().lower()

        if user_input == "done":
            break
        elif user_input in prices:
            price = prices[user_input]
            subtotal += price
            print(f"{user_input.capitalize()} ticket added: ${price:.2f}")
        else:
            print("Invalid ticket type")

    tax = subtotal * 0.08
    total = subtotal + tax

    print("\n--- Receipt ---")
    print(f"Subtotal: ${subtotal:.2f}")
    print(f"Tax (8%): ${tax:.2f}")
    print(f"Total: ${total:.2f}")

# Run the program
movie_ticket_booth()
```

```python
def analyze_grades(grades):
    if not grades:  # Handle empty list case
        return None, None, None

    average = sum(grades) / len(grades)
    highest = max(grades)
    lowest = min(grades)

    return average, highest, lowest
grades = [88, 76, 92, 100, 67]
avg, high, low = analyze_grades(grades)

print(f"Average: {avg:.2f}")
print(f"Highest: {high}")
print(f"Lowest: {low}")
```

```python
def atm_pin_verification():
    CORRECT_PIN = "4321"
    MAX_ATTEMPTS = 3

    for attempt in range(1, MAX_ATTEMPTS + 1):
        entered_pin = input("Enter your 4-digit PIN: ").strip()

        if entered_pin == CORRECT_PIN:
            print("Access granted.")
            return  # Exit the function if successful
        else:
            if attempt < MAX_ATTEMPTS:
                print(f"Wrong ({attempt}/3)")
            else:
                print(f"Wrong ({attempt}/3)")
                print("Card blocked.")

# Run the program
atm_pin_verification()
1111
2222
4321
\
```
